```python
import numpy as np
```

```python
rewards = np.zeros((6, 6))
```

```python
for state in range(6):
    for action in range(6):
        rewards[state][action] = -1
```

```python
q_table = np.zeros((6, 6))
```

```python
rewards[1][5] = 100
```

```python
rewards[4][5] = 100
rewards[5][5] = 100
```

```python
rewards[0][4] = 0
rewards[1][3] = 0
rewards[2][3] = 0
rewards[3][1] = 0
rewards[3][2] = 0
rewards[3][4] = 0
rewards[4][0] = 0
rewards[4][3] = 0
rewards[5][1] = 0
rewards[5][4] = 0
```

```python
rewards
```

```python
gamma = 0.8
```

```python
q_table = np.zeros((6, 6))
```

```python
def train_agent():  
    my_list = []
    for state in range(6):
        for action in range(6):
            if(rewards[state][action] != -1):
                for next_possible_action in range(6):
                        if (rewards[action][next_possible_action] != -1):
                            my_list.append(q_table[action][next_possible_action])
                if len(my_list) > 0:  
                    max_value = max(my_list)
                    print(max_value)
                    q_table[state][action] = int(rewards[state][action] + gamma*max_value)
                    my_list = []
```

```python
train_agent()
q_table
```

```python
for training_iteration in range(30):
    print("Iteration: ", training_iteration)
    train_agent()
    print(q_table)
```
